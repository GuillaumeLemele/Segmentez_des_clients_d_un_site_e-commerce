```python
def shape(df):
    """
    Analyzes the shape of the dataframe: 1) Open it, give some infos 2) Nan values 3) Describe method
    """

    # Head & tail
    df_combined = pd.concat([df.head(), df.tail()], axis=0)
    display(df_combined)
    print("\nNumber of Rows: ", df.shape[0])
    print("Number of Columns: ", df.shape[1])

    for col in df.columns:
            print("Column {}: {}".format(col, df[col].dtype))

    for col in df.columns:
            print("\nUnique values in {}:".format(col))
            print(df[col].unique())

    # N_unique values
    nunique_dict = {col: df[col].nunique() for col in df.columns}
    nunique_df = pd.DataFrame.from_dict(nunique_dict, orient='index', columns=['N_unique values'])
    display(nunique_df)

    # Value counts and missing/duplicated data
    if 'type' in df.columns:
        print("\nValue counts of 'type':")
        print(df['type'].value_counts())
    else:
        print("\n'type' column not found in dataframe.")
    print("\nNumber of missing values:")
    print(df.isna().sum())
    print("\nNumber of duplicated rows:")
    print(df.duplicated().sum())

    # MissingNo
    print("\nMissingNo:")
    print(msno.matrix(df.sample(100, replace = True)))
```

```python
def ana_uni_nom(df, col):
    """
    This function takes a DataFrame and a column name as input and returns basic statistical information about the column.
    
    Parameters:
    df (DataFrame): the DataFrame containing the column of interest.
    col (str): the name of the column to be analyzed.
    
    Returns:
    None: the function prints the following information:
        - Total number of entries in the column.
        - Total number of missing values in the column.
        - Total number of unique values in the column.
        - The mode of the column and its frequency.
        - The value of the mode.
    """
    
    # Get a description of the column from the DataFrame.
    description = df[[col]].describe()
    
    # Get the count of non-null values in the column.
    count = description.loc['count'][0]
    
    # Get the count of null values in the column.
    nan = df[[col]].isna().sum()
    
    # Get the count of unique values in the column.
    unique = description.loc['unique'][0]
    
    # Get the mode of the column.
    top = description.loc['top'][0]
    
    # Get the frequency of the mode.
    freq = description.loc['freq'][0]

    # Print the results.
    print(f"There are {count} entries in the {col} column.")  
    print(f"There are {nan[0]} missing values in the {col} column.")
    print(f"There are {unique} unique values in the {col} column.")
    print(f"The mode of the {col} column is {freq}, which appears {top} times.")
    print(f"The value of the mode in the {col} column is {top}.")
```

```python
def explore_qualitative_variables(df):
    """
    This function takes a DataFrame as input and explores all the qualitative variables in it. 
    For each variable, it prints basic statistical information such as count, number of missing values, 
    number of unique values, mode, and value of the mode.
    
    Parameters:
    df (DataFrame): the DataFrame to be analyzed.
    
    Returns:
    None: the function prints the statistical information for each qualitative variable in the DataFrame.
    """
    
    # Get the names of all columns with qualitative (categorical) data type
    qualitative_cols = df.select_dtypes(include=['object']).columns
    
    # For each qualitative column, analyze the data and print results
    for col in qualitative_cols:
        print(f"\n\nAnalyzing variable: {col}")
        
        ana_uni_nom(df, col)
```

```python
def ana_uni_qtv_er(df, col):
    """
    This function takes a DataFrame and the name of a quantitative variable as input, and computes 
    and prints basic statistics for the variable, including count, number of missing values, minimum value, 
    maximum value, mean, median, first quartile, third quartile, interquartile range (IQR), and standard deviation.
    
    Parameters:
    df (DataFrame): the DataFrame to be analyzed.
    col (str): the name of the quantitative variable to be analyzed.
    
    Returns:
    None: the function prints the statistical information for the variable.
    """
    
    # Compute the basic statistics for the variable using the describe method of DataFrame
    description = df[[col]].describe()
    count = description.loc['count'][0]
    nan = df[[col]].isna().sum()
    min_value = description.loc['min'][0]
    max_value = description.loc['max'][0]
    mean = description.loc['mean'][0]
    median = description.loc['50%'][0]
    q1 = description.loc['25%'][0]
    q3 = description.loc['75%'][0]
    iqr = q3 - q1
    std = description.loc['std'][0]

    # Print the statistical information for the variable
    print(f"There are {count} entries.")  
    print(f"There are {nan[0]} missing values.")
    print(f"The minimum value is {min_value:.2f}.")
    print(f"The maximum value is {max_value:.2f}.")
    print(f"The mean is {mean:.2f}.")
    print(f"The range is {max_value - min_value:.2f}.")
    print(f"The first quartile is {q1:.2f}.")
    print(f"The median is {median:.2f}.")
    print(f"The third quartile is {q3:.2f}.")
    print(f"The interquartile range is {iqr:.2f}.")
    print(f"The standard deviation is {std:.2f}.")
```

```python
def explore_quantitative_variables(df):
    """
    Explore and visualize the quantitative variables in a dataframe.

    Parameters:
    df (pandas.DataFrame): The dataframe to analyze.

    Returns:
    None

    """
    # Select columns with numeric data types
    quantitative_cols = df.select_dtypes(include=['int64', 'float64']).columns

    # Loop through the columns and print the analysis results and visualize the distribution and outliers using boxplot and histogram.
    for col in quantitative_cols:
        print(f"\n\nAnalyzing variable: {col}")
        
        # Call the function ana_uni_qtv_er to analyze the variable
        ana_uni_qtv_er(df, col)

        # Visualize the distribution and outliers using boxplot and histogram
        fig, axs = plt.subplots(ncols=2, figsize=(12, 3))
        sns.boxplot(data=df, x=col, ax=axs[0])
        sns.histplot(data=df, x=col, ax=axs[1])
        axs[0].set_title(f'Boxplot of: {col}')
        axs[0].set_xlabel(col)
        axs[1].set_title(f'Histogram of: {col}')
        axs[1].set_xlabel(col)
        plt.show()
```

```python
def remove_outlier(input_df, colum_name):
    """
    Removes outliers from a given column in a Pandas DataFrame.

    Parameters:
        input_df (pd.DataFrame): The input DataFrame.
        colum_name (str): The name of the column to remove outliers from.

    Returns:
        pd.DataFrame: A new DataFrame with the outliers removed.

    """
    # Calculate the first and third quartile (Q1 and Q3), and the interquartile range (IQR)
    q1 = input_df[colum_name].quantile(0.05)
    q3 = input_df[colum_name].quantile(0.95)
    iqr = q3-q1     

    # Calculate the threshold for low and high outliers
    threshold_low_outliers  = q1-1.5*iqr
    threshold_high_outliers = q3+1.5*iqr

    # Get the indices of the outliers
    index_outliers = input_df.loc[(input_df[colum_name] < threshold_low_outliers) | (input_df[colum_name] > threshold_high_outliers)].index

    # Drop the rows containing the outliers
    input_df = pd.DataFrame(input_df.drop(index_outliers.to_list(), axis=0, inplace=True))

    return input_df
```

## Customer et seller state (& Feature Engineering, + cleaning)

```python
# Je pars du principe avec un unique pour les customers States que les consommateurs ne changent pas d'addresse, car je selectionne la première valeurs.
# Pour les types de paiement, j'ai sélectionner un nunique pour faire la somme de chaque type de paiement, qui est une valeur unique dans sa variable d'origine.
# Nous créons une variable ou  1 si customer_state est égale à seller_state et 0 sinon.
# Nous dropons seller state et customer state.
# J'ai un biais sur ma variable crée car 701 valeurs nan, et j'ai mis non 0, si on ne sais pas, pe mettre 2? (verif ville, si nan seller state y sont sur city mais de coté pour le moment)

state = dff.groupby('customer_unique_id').agg({
            'seller_state': lambda x: x.unique()[0],
            'customer_state': lambda x: x.unique()[0],
            }).reset_index()

state['same_state'] = state.apply(lambda x: 1 if x['customer_state'] == x['seller_state'] else 0, axis=1)
state.drop(['seller_state', 'customer_state'], axis=1, inplace=True)

df_grouped = df_grouped.merge(state, on='customer_unique_id', how='left')
df_grouped.head()
```

## Temps de livraison des commandes

```python
# Nous décidons, au lieu de tronquer les valeurs, de garder la précision des heures au sein de nos nombres décimaux. 
# Nous décidons de remplir les NAN, soit les commandes annulés, ou non livrés, par 0.

df_grouped["Temps_livraison"] = dff['order_delivered_customer_date'] - dff['order_purchase_timestamp'].dt.round('1d')
df_grouped["Temps_livraison"] = df_grouped["Temps_livraison"].dt.round('1d')
df_grouped["Temps_livraison"] = (dff['order_delivered_customer_date'] - dff['order_purchase_timestamp']).dt.total_seconds() / 86400.0
df_grouped["Temps_livraison"] = df_grouped["Temps_livraison"].round(1)

df_grouped["Temps_livraison"].fillna(0, inplace=True)


shape(df_grouped)
```

```python
# Décomposation saisonale avec la librairie Statmodel
# Le graphique obtenu montre la tendance au fil des mois ainsi que la composante saisonnière, permettant ainsi de mieux comprendre l'évolution des ventes au fil du temps.

# On réinstencie autrement df_grouped_date
df_grouped_date = df_grouped

# On commence par grouper par mois/année en créant une nouvelle colonne 'purchase_month_year'
df_grouped_date['achats_mois'] = df_grouped['order_purchase_timestamp'].dt.to_period('M')

# On calcule le nombre de commandes par mois/année, on elève ainsi les duplicas. 
count_by_month_year = df_grouped_date.groupby('achats_mois').size()

# On convertit l'index en type datetime
count_by_month_year.index = count_by_month_year.index.astype('datetime64[ns]')

# On trie les données par date
count_by_month_year.sort_index(inplace=True)

# On effectue la décomposition saisonnière en utilisant une période de 10 mois dû à notre dataset
result = seasonal_decompose(count_by_month_year, model='additive', period=10)

# On crée un graphique avec deux quatre-graphiques : le premier pour la tendance au fil des mois et le second pour la décomposition saisonnière (le premier étant les données observées de base et dernier étant le bruit)
fig, ax = plt.subplots(4,1,figsize=(12,10))

# Graphique pour les données observés
sns.lineplot(data=result.observed, x=result.observed, y=result.observed, ax=ax[0])
ax[0].set_title('rappel : données observés non décomposées')

# Graphique pour la tendance
sns.lineplot(data=result.trend, x=result.trend.index, y=result.trend, ax=ax[1])
ax[1].set_title('Tendance au fil des mois')

# Graphique pour la décomposition saisonnière
sns.lineplot(data=result.seasonal, x=result.seasonal.index, y=result.seasonal, ax=ax[2])
ax[2].set_title('Décomposition saisonale')

# Graphique pour le bruit
sns.lineplot(data=result.resid, x=result.resid, y=result.resid, ax=ax[3])
ax[3].set_title('Bruit')



# Réglage des marges pour éviter les chevauchements
plt.tight_layout()

# Affichage du graphique
plt.show()
```

**1) Nous n'avons pas d'informations si ces outliers sont des individus atypiques ou des valeurs aberrantes.**  
**2) Si ce sont des valeurs aberrantes => il est normal que le pourcentage alors soit plus haut que 3%, ce qui après suppression de ces valeurs, nous donne exactement 3%, chiffre qui nous est donné par la société qui nous a confiée ce travail. Nous allons donc considérer qu'un pourcentage plus haut que 3% est une erreur au vu des informations commandités par l'entreprise.**   
**3) Il serait par ailleurs nécessaire de faire remonter l'information et de faire vérifier manuellement par la société ces valeurs si elle n'a pas les informations supposées, ou de nous préciser leurs informations; afin de valider si ce sont bien des erreurs.**  
**4) Si ce sont des valeurs atypiques, ils sont à regroupé dans un "cluster", ou groupe spécial de client "VIP" avec un traitement préférentiel par rappport à tous les autres types de client (ou alors ce sont des personnes morales derrières les individus), à traiter différement vu la valeur monétaire des individus. Il n'est pas forcément significatif, après traitement de l'influence des outliers, de faire rentrer ces individus dans nos modèles de partionnement, à vérifier pour plus tard.**
